```python
dimension = 2
num_layers = 7
dt = 1/num_layers

'''dataset_spiral_train = make_moons(n_samples=100, noise=0.1, random_state=0)
mu_spiral, std_spiral = dataset_spiral_train[0].mean(axis=0), dataset_spiral_train[0].std(axis=0)
dataset_spiral_train = ((dataset_spiral_train[0] - mu_spiral) / std_spiral), dataset_spiral_train[1]
dataset_spiral_test = make_moons(n_samples=80, noise=0.1, random_state=1)
dataset_spiral_test = ((dataset_spiral_test[0] - mu_spiral) / std_spiral), dataset_spiral_test[1]'''

dataset_donut_train = make_circles(n_samples=100, noise=0.1, factor=0.2, random_state=1)
mu_donut, std_donut = dataset_donut_train[0].mean(axis=0), dataset_donut_train[0].std(axis=0)
dataset_donut_train = ((dataset_donut_train[0] - mu_donut) / std_donut), dataset_donut_train[1]
dataset_donut_test = make_circles(n_samples=80,noise=0.1, factor=0.2, random_state=2)
dataset_donut_test = ((dataset_donut_test[0] - mu_donut) / std_donut), dataset_donut_test[1]


# Instantiate initial control vector
control_0 = []
for i in range(num_layers):
    K = [1,0,0,1]
    be = [0,0]
    Dt = [dt]
    control_0 += K+be+Dt
W = [1,0,0,1]
mu = [0,0]
control_0 += W+mu

control_0 = np.array(control_0)


alpha=0.15
#X, y = dataset_spiral_train
X, y = dataset_donut_train

plot_datasets(dataset_donut_train, dataset_donut_test)
```

```python
def plot_region(X, y, clf, model=""):
    fig, ax = plt.subplots(1, 1, figsize=(5*1, 5))
    h=0.1
    x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
    y_min, y_max = X[:,1].min()-1, X[:,1].max()+1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, cmap="RdYlBu", alpha=0.5)
    ax.scatter(X[:,0], X[:,1], c=y, cmap="RdYlBu")
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(model, fontsize=12)
    ax.set_ylabel("Prediction", fontsize=12)


plot_region(dataset_donut_test[0], dataset_donut_test[1], mlp_clf_spiral, model="NN")
```

### ResNet-New

```python
def dynamic(control, X, alpha=1):
    O = []
    T = []
    for k in range(X.shape[0]):
        x0 = X[k]
        xn = x0.copy()
        x = [xn]
        for i in range(num_layers):
            ix = dimension*(dimension+1)*i
            jx = ix+dimension*dimension
            zx = jx+dimension
            K = control[ix:jx]
            be = control[jx:zx]
            KK = np.array([
                [K[0], K[1]],
                [K[2], K[3]]
            ])
            xn = (xn + alpha*dt*activation_function(KK@xn + be)).copy()
            xn = xn
            x.append(xn)

        ix = dimension*(dimension+1)
        jx = -ix+dimension*dimension
        W = control[-ix:jx]
        mu = control[jx:]
        WW = np.array([
            [W[0], W[1]],
            [W[2], W[3]]
        ]) 
        xf = (WW@xn + mu).copy()
        x.append(xf)
        T.append(x)
        sxf = np.exp(xf)
        sxf /= sum(sxf)
        o = sxf[1]
        O.append(o)
    O = np.array(O)
    return O, T


def obj(control, X, y, alpha=1):
    O, _ = dynamic(control, X, alpha)
    L = y[:O.shape[0]]
    return 0.5*sum((O-L)*(O-L))


def get_cls(control):
        ix = dimension*(dimension+1)
        jx = -ix+dimension*dimension
        W = control[-ix:jx]
        mu = control[jx:]
        WW = np.array([
            [W[0], W[1]],
            [W[2], W[3]]
        ]) 
        return WW, mu


dimension = 2
num_layers = 5
#activation_function = lambda x: np.tanh(x)
activation_function = lambda x: 1/(1 + np.exp(-x))
dt = 1/num_layers

control_0 = []
for i in range(num_layers):
    K = [1,0,0,1]
    be = [0,0]
    control_0 += K+be
W = [1,0,0,1]
mu = [0,0]
control_0 += W+mu

control_0 = np.array(control_0)

alpha=0.15
X, y = dataset_donut_train

sol_donut = minimize(obj, x0=control_0, args=(X, y, alpha))

optimal_control_donut = sol_donut.x
O_donut, T_donut = dynamic(optimal_control_donut, X, alpha)
WW_donut, mu_donut = get_cls(optimal_control_donut)
```

```python

plot_iterations(num_layers, T_donut, dataset_donut_train[1], WW_donut, mu_donut)
```

```python
predicted_class = (O_donut >= 0.5).astype(int)
accuracy = accuracy_score(y, predicted_class)
print (f'ResNet Accuracy: {accuracy}')
```
